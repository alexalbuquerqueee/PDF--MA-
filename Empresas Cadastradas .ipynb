```python
#Importando bibiliotecas 

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#Carregand DataSets

dados = pd.read_excel("C:/Users/alexa/Downloads/Empresas Cadastradas.xlsx")
display(dados)
```

```python
#Excluindo colunas sem valores
dados.dropna(axis=1)
```

```python
#Plotando Gricos e Tamanhos
 
plt.figure(figsize = (10,5))
plt.plot(dados.ANO, dados.EMPRESAS, label = "Empresas Cadastradas", lw = "3")
plt.plot(dados.ANO, dados.TOTAL, label = "Total de Empresas Cadastradas", lw = "3", color = "red")
plt.legend() 
plt.title("Empresas Cadastradas")
```

```python
#Variao do nmero de Empresas ao longo do Tempo

plt.plot(dados.ANO, dados.EMPRESAS, label = "Empresas Cadastradas", lw = "3")
plt.ylabel("Nmero de Empresas")
plt.xlabel("Anos")
plt.title("Variao do Nmero de Empresas")


plt.yticks([0,20,40,60,80,100,120])
```

```python
#Variao do nmero de Empresas Cadastradas no TOTAl

plt.plot(dados.ANO, dados.TOTAL, label = "Total de Empresas Cadastradas", lw = "3", color = "red")
plt.ylabel("Nmero de Empresas")
plt.xlabel("Anos")
plt.title ("Total de Empresas Cadastradas")
```
